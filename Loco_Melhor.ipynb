```python
df=pd.read_csv('gdrive/My Drive/DelhiNCR Restaurants.csv')
df1 = df
df
```

```python
df = df.drop(["Known_For2","Known_For22","Delivery_Rating","Delivery_Rating_Count","Restaurant_Name"],axis=1)
df
```

```python
df["Dining_Rating"].unique()
```

```python
df.loc[(df["Dining_Rating"] < 4.3) | (df["Dining_Review_Count"] < 600),"Label"] = "Bad"
df
```

```python
df["Label"] = df["Label"].fillna("Good")
```

```python
df2 = df
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
df['Locality'] = df['Locality'].str.split(',').str[0] 
df
```

```python
columns = ["Locality","Dining_Rating","Dining_Review_Count","Latitude","Longitude"]
x = df[columns]
y = df["Label"]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
x["Locality"] = le.fit_transform(x["Locality"])
```

```python
x["Locality"].unique()
```

```python
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0).fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
